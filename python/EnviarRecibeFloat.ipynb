# Comunicacion con el arduino por el puerto Serial para numeros en coma flotante

```python
import serial
import time
import struct

ser = serial.Serial('COM3', 9600, timeout=1)
point1 = [1.7, 1.7, 1.7]
point2 = [8.7, 8.7, 8.7]
point3 = [20.7, 20.7, 20.7]
point4 = [2.7, 1.7, 1.7]
point5 = [1.7, 2.7, 2.7]
point6 = [5.7, 8.7, 9.7]
point7 = [20.7, 21.7, 21.7]
point8 = [20.7, 25.7, 20.7]
point9 = [20.7, 20.7, 25.7]

points = [ point1, point2, point3, point4, point5, point6, point7, point8, point9 ]
#3, false, 0.1, 3

def setUpStreamKeans(numKlusters, forget, learningR, numDimensions):

    data = str(numKlusters) + "\n"
    data = str(forget) + "\n"
    data = str(learningR) + "\n"
    data = str(numDimensions) + "\n"
    ser.write(data.encode())

setUpStreamKeans(3, 0, 0.1, 3)

def writePoint(PointToWrite):
    lenghtPoint = len(PointToWrite)
    for indexPoint in range(lenghtPoint):
        data= PointToWrite[indexPoint]
        data = str(data) + "\n"
        ser.write(data.encode())

i = 0
while True:
    if i > 8:
        i = 0
    writePoint(points[i])
    time.sleep(0.2)
    print(ser.readline())
    i = i + 1



```
